```python
import time
def isSubsetSum(set, n, sum):
    subset = [[False for _ in range(sum + 1)] for _ in range(n + 1)]

    for i in range(n + 1):
        subset[i][0] = True

    for i in range(1, sum + 1):
        subset[0][i] = False

    for i in range(1, n + 1):
        for j in range(1, sum + 1):
            if j < set[i - 1]:
                subset[i][j] = subset[i - 1][j]
            else:
                subset[i][j] = subset[i - 1][j] or subset[i - 1][j - set[i - 1]]

    if not subset[n][sum]:
        return []

    selected_elements = []
    i, j = n, sum
    while i > 0 and j > 0:
        if subset[i][j] != subset[i - 1][j]:
            selected_elements.append(set[i - 1])
            j -= set[i - 1]
        i -= 1

    return selected_elements

# Driver code
if __name__ == '__main__':
    n_values = list(range(50, 1001, 50))
    for n in n_values:
        start_time = time.time()
        avg_time=0
        target_sum = 200  # Tổng cần kiểm tra
        random_set = generate_random_set(n)  # Sinh ngẫu nhiên tập set
        print("Random set:", random_set)
        subset = isSubsetSum(random_set, n, target_sum)
        if subset:
            print("Subset with given sum:", subset)
        else:
            print("No subset with given sum")
        end_time = time.time()
        avg_time += end_time - start_time
        print("Avg time ",avg_time)
```
